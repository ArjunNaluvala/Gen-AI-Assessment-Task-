**Task 1: Replace Hardcoded Prompts**

```python
# Original (BAD): def explain_topic(topic): return f"Explain {topic} in simple terms for beginners"
# Requirement: No f-strings, use PromptTemplate, must be reusable

# Import the necessary class for creating prompt templates
from langchain_core.prompts import PromptTemplate

# Define a reusable PromptTemplate with input_variables
# This template takes a 'topic' as input and formats a string for explanation.
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Reusable function using the template (no f-strings, no hardcoding)
# This function encapsulates the prompt generation, making it easy to reuse.
def explain_topic(topic: str) -> str:
    """Generate a beginner-friendly explanation prompt for any topic."""
    return explain_template.format(topic=topic)

# Test the template with different topics to demonstrate reusability
print("Testing explain_topic function")

# Example 1: Explaining Artificial Intelligence
artificial_intelligence_prompt = explain_topic("Artificial Intelligence")
print(artificial_intelligence_prompt)

# Example 2: Explaining Photosynthesis
photosynthesis_prompt = explain_topic("Photosynthesis")
print(photosynthesis_prompt)

# Example 3: Explaining Quantum Physics
quantum_physics_prompt = explain_topic("Quantum Physics")
print(quantum_physics_prompt)
```

**Task 3: Prompt Variations Engine**

```python
# Define various PromptTemplate instances for different prompt styles.
# Each template is designed to generate a specific type of prompt based on a given 'topic'.

# Template for a teaching style prompt: explains a topic step by step.
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Template for an interview style prompt: asks questions about a topic.
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Template for a storytelling style prompt: explains a topic as a narrative.
storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Reusable functions to format each prompt template with a specific topic.
# These functions abstract the formatting logic, making it easier to generate prompts.

def prompt_teaching(topic):
    """Generates a teaching-style prompt for the given topic."""
    return teaching_template.format(topic=topic)

def prompt_interview(topic):
    """Generates an interview-style prompt for the given topic."""
    return interview_template.format(topic=topic)

def prompt_storytelling(topic):
    """Generates a storytelling-style prompt for the given topic."""
    return storytelling_template.format(topic=topic)

# --- Prompt Variations Engine Test ---
# This section demonstrates how to use the defined templates and functions
# to generate different styles of prompts for a single topic.

topic_task3 = "Machine Learning"

print("[Teaching Style]")
# Generate and print a teaching-style prompt
print(prompt_teaching(topic_task3))

print("\n[Interview Style]")
# Generate and print an interview-style prompt
print(prompt_interview(topic_task3))

print("\n[Storytelling Style]")
# Generate and print a storytelling-style prompt
print(prompt_storytelling(topic_task3))
```

**Task 4: ChatPromptTemplate System**

```python
# Import necessary modules for ChatPromptTemplate
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

# System role → defines behavior | User role → dynamic input
# Supported roles: teacher, interviewer, motivator

# Define system behavior descriptions per role. These will be injected into the SystemMessagePromptTemplate.
# Storing them in a dictionary prevents hardcoding role descriptions directly within the function.
role_descriptions = {
    "teacher":     "You are an expert teacher. Explain concepts clearly with examples and step-by-step guidance.",
    "interviewer": "You are a professional interviewer. Ask insightful, thought-provoking questions about the topic.",
    "motivator":   "You are an enthusiastic motivator. Inspire the user to learn and explore the topic with energy."
}

def build_chat_prompt(role: str, topic: str):
    """
    Builds a ChatPromptTemplate dynamically based on a specified role and topic.

    This function constructs a list of message objects, typically consisting of
    a SystemMessage and a HumanMessage, which can be used to prime a chatbot
    or language model for a specific interaction.

    Args:
        role (str): The desired role for the AI ('teacher', 'interviewer', 'motivator').
                    This dictates the persona and behavior of the AI.
        topic (str): The subject matter the user wants to discuss or learn about.

    Returns:
        list: A list of message objects (e.g., [SystemMessage, HumanMessage])
              formatted according to the chosen role and topic.

    Raises:
        ValueError: If an unsupported role is provided, ensuring that only
                    predefined roles are used.
    """
    # Validate the provided role against the list of supported roles.
    if role not in role_descriptions:
        raise ValueError("Role must be one of: teacher, interviewer, motivator")

    # Create a SystemMessagePromptTemplate that dynamically injects the role's behavior.
    # This sets the context and persona for the AI.
    system_template = SystemMessagePromptTemplate.from_template(
        "{role_description}"
    )

    # Create a HumanMessagePromptTemplate for the user's input, dynamically including the topic.
    # This is the actual query or instruction from the user.
    human_template = HumanMessagePromptTemplate.from_template(
        "Please help me understand {topic}."
    )

    # Combine the system and human message templates into a single ChatPromptTemplate.
    # This represents the full conversation turn.
    chat_prompt = ChatPromptTemplate.from_messages([system_template, human_template])

    # Format the combined prompt with the actual role description and topic.
    # This generates the final list of message objects ready for a chat model.
    formatted = chat_prompt.format_messages(
        role_description=role_descriptions[role],
        topic=topic
    )
    return formatted

# --- Test Cases for the ChatPromptTemplate System ---
print("Task 4: ChatPromptTemplate System")

# Define a list of test scenarios, each with a role and a topic.
test_roles = [
    {"role": "teacher",     "topic": "Neural Networks"},
    {"role": "interviewer", "topic": "Deep Learning"},
    {"role": "motivator",   "topic": "Artificial Intelligence"},
]

# Iterate through each test scenario to build and print the chat prompts.
for entry in test_roles:
    # Build the chat prompt using the defined role and topic.
    messages = build_chat_prompt(**entry)
    print("\n[Role: {} | Topic: {}]".format(entry['role'].upper(), entry['topic']))
    # Print each message within the generated chat prompt, indicating its type and content.
    for msg in messages:
        print("  {}: {}".format(msg.type.upper(), msg.content))

print("\nChatPromptTemplate System setup complete and tested")
```

**Task 5: Input Validation Layer**

```python
# audience ∈ [beginner, intermediate, expert]
# tone ∈ [formal, casual, fun]
# Strategy: raise ValueError for invalid inputs

# Define valid options as constants (not hardcoded inside function logic)
VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES = ["formal", "casual", "fun"]

# Default fallback values for audience and tone
DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE = "casual"

def validate_inputs(audience, tone, use_defaults=False):
    """
    Validates audience and tone inputs against predefined lists.
    If inputs are invalid, it either assigns default values (if use_defaults is True)
    or raises a ValueError (if use_defaults is False).

    Args:
        audience (str): The requested audience for the prompt.
        tone (str): The requested tone for the prompt.
        use_defaults (bool): If True, invalid inputs will be replaced by defaults.
                             If False, invalid inputs will raise a ValueError.

    Returns:
        tuple: A tuple containing the validated (or defaulted) audience and tone.

    Raises:
        ValueError: If an input is invalid and use_defaults is False.
    """
    # Normalize inputs: convert to lowercase and remove leading/trailing whitespace
    audience = audience.lower().strip()
    tone = tone.lower().strip()

    # Validate audience input
    if audience not in VALID_AUDIENCES:
        # If validation fails, check if default values should be used
        if use_defaults:
            print("  [WARNING] Invalid audience '{}'. Using default: '{}'".format(audience, DEFAULT_AUDIENCE))
            audience = DEFAULT_AUDIENCE  # Assign default audience
        else:
            # If defaults are not to be used, raise an error indicating invalid input
            raise ValueError("Invalid audience '{}'. Must be one of: {}".format(audience, VALID_AUDIENCES))

    # Validate tone input
    if tone not in VALID_TONES:
        # If validation fails, check if default values should be used
        if use_defaults:
            print("  [WARNING] Invalid tone '{}'. Using default: '{}'".format(tone, DEFAULT_TONE))
            tone = DEFAULT_TONE  # Assign default tone
        else:
            # If defaults are not to be used, raise an error indicating invalid input
            raise ValueError("Invalid tone '{}'. Must be one of: {}".format(tone, VALID_TONES))

    # Return the validated (or defaulted) audience and tone
    return audience, tone


print("Task 5: Input Validation Layer")

# Test 1: Valid inputs - Expects successful validation and original values
print("\n[Test 1 - Valid inputs]")
result = validate_inputs("beginner", "fun")
print("  Validated: audience={}, tone={}".format(result[0], result[1]))

# Test 2: Invalid inputs with defaults - Expects warnings and default values
print("\n[Test 2 - Invalid inputs with defaults]")
result = validate_inputs("child", "dramatic", use_defaults=True)
print("  Validated: audience={}, tone={}".format(result[0], result[1]))

# Test 3: Invalid inputs raising ValueError - Expects a ValueError to be caught
print("\n[Test 3 - Invalid inputs raising ValueError]")
try:
    validate_inputs("child", "dramatic", use_defaults=False)
except ValueError as e:
    print("  Caught ValueError: {}".format(e))
```

**Task 6: Prompt Generator App**

```python
# Integrates: validation + style selection + PromptTemplate
# Pipeline: User Input → Validation → Template → Output

# Style-specific PromptTemplates (modular, separate from logic)
# This dictionary stores PromptTemplate instances for different prompt styles.
# Each template expects 'topic', 'audience', and 'tone' as input variables.
style_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style, step by step."
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Generate interview questions about {topic} for {audience} in a {tone} tone."
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style."
    )
}

def generate_prompt(topic: str, audience: str, tone: str, style: str) -> str:
    """
    Generates a formatted prompt based on the specified topic, audience, tone, and style.

    This function integrates input validation and dynamic template selection to create
    a robust prompt generation system.

    Args:
        topic (str): The subject matter for the prompt.
        audience (str): The target audience (e.g., 'beginner', 'expert').
        tone (str): The desired tone (e.g., 'formal', 'fun').
        style (str): The prompt style to use (e.g., 'teaching', 'interview', 'storytelling').

    Returns:
        str: The fully formatted prompt string.

    Raises:
        ValueError: If an invalid style is provided.
    """
    # Step 1: Validate inputs using Task 5's validation function
    # The validate_inputs function handles normalization and can use defaults or raise errors.
    audience, tone = validate_inputs(audience, tone, use_defaults=True)

    # Step 2: Validate the requested style
    # Ensure the style exists in our predefined style_templates dictionary.
    style = style.lower().strip()
    if style not in style_templates:
        # If the style is invalid, raise a ValueError with available options.
        raise ValueError("Invalid style '{}'. Must be one of: {}".format(style, list(style_templates.keys())))

    # Step 3: Select the appropriate template and format the prompt
    # Retrieve the PromptTemplate instance corresponding to the validated style.
    selected_template = style_templates[style]
    # Format the template with the validated topic, audience, and tone.
    final_prompt = selected_template.format(topic=topic, audience=audience, tone=tone)

    return final_prompt


print("Task 6: Prompt Generator App")

# Example from assignment: Generate a storytelling style prompt for Neural Networks
example = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)
print("Example Output:\n  → {}".format(example))

# Additional test cases to demonstrate flexibility and reusability
additional_tests = [
    {"topic": "Python",          "audience": "intermediate", "tone": "casual",  "style": "teaching"},
    {"topic": "Data Science",    "audience": "expert",       "tone": "formal",  "style": "interview"},
    {"topic": "Deep Learning",   "audience": "beginner",     "tone": "fun",     "style": "storytelling"},
]

print("\nAdditional Tests:")
for test in additional_tests:
    # Unpack the dictionary to pass arguments to generate_prompt
    result = generate_prompt(**test)
    print("  → {}".format(result))
```

**Task 7: Template Reusability Test**

```python
# ONE template → 5 different inputs → 5 different outputs
# Goal: Same structure, completely different content

# Single reusable master template
master_template = PromptTemplate(
    input_variables=["topic", "audience", "tone", "style"],
    template=(
        "You are a {style} expert. "
        "Explain {topic} to a {audience}-level learner in a {tone} tone."
    )
)

# 5 completely different input combinations
reusability_inputs = [
    {"topic": "Machine Learning",   "audience": "beginner",     "tone": "fun",     "style": "teacher"},
    {"topic": "Blockchain",         "audience": "intermediate", "tone": "casual",  "style": "journalist"},
    {"topic": "Quantum Computing",  "audience": "expert",       "tone": "formal",  "style": "researcher"},
    {"topic": "Natural Language Processing", "audience": "beginner", "tone": "friendly", "style": "mentor"},
    {"topic": "Reinforcement Learning", "audience": "intermediate", "tone": "technical", "style": "engineer"},
]

print("Task 7: Template Reusability Test")
# Display the template being used for clarity
print("Template Used: '{}'\n".format(master_template.template))
print("5 Different Outputs from the SAME Template")

for i, inputs in enumerate(reusability_inputs, start=1):
    # Same template, different data each time
    output = master_template.format(**inputs)
    print("\n[Run {}]".format(i))
    # Display the inputs used for each run
    print("  Input : topic={}, audience={}, tone={}, style={}".format(
        inputs['topic'], inputs['audience'], inputs['tone'], inputs['style']
    ))
    # Display the generated output
    print("  Output: {}".format(output))

# Confirm that reusability has been successfully demonstrated
print("\nReusability confirmed: Same template structure, 5 unique outputs.")
```

**Pipeline Flow Code**

```python
def full_pipeline(topic, audience, tone, style):
    """
    Executes the complete prompt generation pipeline, including input display,
    validation, template selection, and final prompt generation.

    Args:
        topic (str): The subject matter for the prompt.
        audience (str): The target audience (e.g., 'beginner', 'expert').
        tone (str): The desired tone (e.g., 'formal', 'fun').
        style (str): The prompt style to use (e.g., 'teaching', 'interview', 'storytelling').

    Returns:
        str: The final generated prompt string.
    """
    print("=" * 50)
    print("MINI PROMPT ENGINE – Full Pipeline")
    print("=" * 50)
    # Step 1: User Input - Display the initial inputs for clarity.
    print("[Step 1] User Input   → topic={}, audience={}, tone={}, style={}".format(topic, audience, tone, style))

    # Step 2: Validation - Validate the audience and tone using the previously defined function.
    # It uses defaults if invalid inputs are provided.
    validated_audience, validated_tone = validate_inputs(audience, tone, use_defaults=True)
    print("[Step 2] Validation   → audience={}, tone={}".format(validated_audience, validated_tone))

    # Step 3: Template selection + formatting - Generate the final prompt string.
    final_prompt = generate_prompt(topic, validated_audience, validated_tone, style)
    print("[Step 3] Template     → Style '{}' template selected".format(style))

    # Step 4: Output - Display the final generated prompt.
    print("[Step 4] Final Output → {}".format(final_prompt))
    print("=" * 50)
    return final_prompt


# Run the full pipeline with an example to demonstrate its functionality.
full_pipeline(
    topic="Artificial Intelligence",
    audience="beginner",
    tone="fun",
    style="storytelling"
)
```
